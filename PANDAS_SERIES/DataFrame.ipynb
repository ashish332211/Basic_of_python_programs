```python
#Find all the final matches
mask = ipl['MatchNumber'] == 'Final'
ipl[mask]
```

```python
#Find all the final winners
mask = ipl['MatchNumber'] == 'Final'
new_df = ipl[mask]
new_df[['Season','WinningTeam']]
```

```python
#in one line
ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
#How many super over finishes have occured
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
#How many matches has csk won in kolkata
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
#How many time csk played final
ipl[(ipl['MatchNumber'] == 'Final') & ((ipl['Team1']=='Chennai Super Kings') | (ipl['Team2']=='Chennai Super Kings'))].shape[0]
```

```python
#How many times Toss winner is the Match Winner
ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]
```

```python
#Toss winner is the Match Winner Percentage
(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python
#Movies with rating higher than 8 and votes>10000
movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)].shape[0]
```
